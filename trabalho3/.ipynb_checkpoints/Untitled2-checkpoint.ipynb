```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier,RandomForestClassifier
from sklearn.metrics import classification_report


dataset = pd.read_csv('am-T3-dados/credtrain.txt',delimiter="\t", header = None, names = ["ESCT", "NDEP", "RENDA", "TIPOR","VBEM","NPARC","VPARC","TEL","IDADE","RESMS","ENTRADA","CLASSE"])
dataset
cols = dataset.shape[1]

X = dataset.iloc[:,0:cols-1]
y = dataset['CLASSE']
scaler = StandardScaler()
X_s_normalize = scaler.fit_transform(X)
x_train,x_test,y_train,y_test = train_test_split(X_s_normalize,y,test_size=0.2)
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
import numpy
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.constraints import MaxNorm
from sklearn.metrics import confusion_matrix



# input_dim = x_train.shape[1]



def create_model(neurons=1):
    # default values
#     activation='relu' # or linear
#     dropout_rate=0.0 # or 0.2
#     init_mode='uniform'
#     weight_constraint=0 # or  4
#     optimizer='adam' # or SGD
#     lr = 0.01
#     momemntum=0
    # create model
    model = Sequential()
    L1 = Dense(neurons, input_dim=11, kernel_initializer='uniform', activation='sigmoid', kernel_constraint=MaxNorm(4))
    model.add(L1)
    L2 = Dense(1, activation='sigmoid')
    model.add(L2)
    
    model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['acc'])
    return model


# create model
model = KerasClassifier(build_fn=create_model, batch_size=10, epochs=100) 
# use verbose=0 if you do not want to see progress

########################################################
# Use scikit-learn to grid search 
# activation =  ['relu', 'tanh', 'sigmoid', 'hard_sigmoid', 'linear'] # softmax, softplus, softsign 
# momentum = [0.0, 0.2, 0.4, 0.6, 0.8, 0.9]
# learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
# dropout_rate = np.arange(0.0,0.9,0.1)
# weight_constraint= np.arange(1,5)
# init = ['uniform', 'lecun_uniform', 'normal', 'zero', 'glorot_normal', 'glorot_uniform', 'he_normal', 'he_uniform']
# optimizer = [ 'SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam']
##############################################################
# grid search epochs, batch size
epochs = [10, 100] # add 50, 100, 150 etc
batch_size = [1000, 5000] # add 5, 10, 20, 40, 60, 80, 100 etc
neurons = np.arange(1, 30, 1).tolist()

param_grid = dict(neurons=neurons)
##############################################################


grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1)
grid_result = grid.fit(x_train,y_train)
history  = grid_result.best_estimator_.model.history
y_pred = grid.predict(x_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
metrics = history.history
metrics.keys()
plt.figure(figsize=(14,7))

plt.plot(metrics['acc'])
plt.plot(metrics['loss'])
plt.title('Curvas de Treinamento')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Precisão', 'Função de Custo'], loc='upper left')
plt.show();
```

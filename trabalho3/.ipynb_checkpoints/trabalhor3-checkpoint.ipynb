# Aprendizado de Comitês

### Para executar o treinamento utilizando aprendizado de comites precisa ter aba base credtrain que esta no formato txt . read_csv premite ler a base e setar nomes para as colunas extraidas . Apos esse passo, utiliza se normalizacao afim de alterar os valores das colunas numéricas no conjunto de dados em uma escala comum, isso sem alterar ou distorcer os intervalos de valores. Utilizar esse  método é importante apenas quando os parâmetros tiverem intervalos muito diferentes.

### Apos isso utiliza se o metodo train_test_split para separar os dados de treinamento dos dados de test

```python
import numpy as np
import pandas as pd# Aprendizado de Comitês

import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier,RandomForestClassifier
from sklearn.metrics import classification_report


dataset = pd.read_csv('am-T3-dados/credtrain.txt',delimiter="\t", names = ["ESCT", "NDEP", "RENDA", "TIPOR","VBEM","NPARC","VPARC","TEL","IDADE","RESMS","ENTRADA","CLASSE"])
dataset
cols = dataset.shape[1]

X = dataset.iloc[:,0:cols-1]
y = dataset['CLASSE']


scaler = StandardScaler()
X_s_normalize = scaler.fit_transform(X)

x_train,x_test,y_train,y_test = train_test_split(X_s_normalize,y,test_size=0.2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier,BaggingClassifier,RandomForestClassifier
from sklearn.metrics import classification_report


dataset = pd.read_csv('am-T3-dados/credtrain.txt',delimiter="\t", header = None, names = ["ESCT", "NDEP", "RENDA", "TIPOR","VBEM","NPARC","VPARC","TEL","IDADE","RESMS","ENTRADA","CLASSE"])
dataset
cols = dataset.shape[1]

X = dataset.iloc[:,0:cols-1]
y = dataset['CLASSE']
scaler = StandardScaler()
X_s_normalize = scaler.fit_transform(X)
x_train,x_test,y_train,y_test = train_test_split(X_s_normalize,y,test_size=0.2)
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
import numpy
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.constraints import MaxNorm
from sklearn.metrics import confusion_matrix

def create_model(neurons=1, activation='sigmoid'):
    model = Sequential()
    L1 = Dense(neurons, input_dim=11, kernel_initializer='uniform', activation=activation, kernel_constraint=MaxNorm(4))
    model.add(L1)
    L2 = Dense(1, activation='sigmoid')
    model.add(L2)
    
    model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['acc'])
    return model


# create model
model = KerasClassifier(build_fn=create_model, batch_size=10, epochs=100) 
# use verbose=0 if you do not want to see progress

########################################################
# Use scikit-learn to grid search 
activation =  ['relu', 'sigmoid', 'softmax']
# momentum = [0.0, 0.2, 0.4, 0.6, 0.8, 0.9]
# learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
dropout_rate = np.arange(0.0,0.9,0.1)
# optimizer = [ 'SGD','Adam', 'Adamax', 'Nadam']
##############################################################
# grid search epochs, batch size
epochs = [100,200] # add 50, 100, 150 etc
batch_size = [100,150] # add 5, 10, 20, 40, 60, 80, 100 etc
neurons = np.arange(1,10,2).tolist()

param_grid = dict(neurons=neurons,activation=activation,epochs=epochs,batch_size=batch_size)
##############################################################

grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1)
grid_result = grid.fit(x_train,y_train)
history  = grid_result.best_estimator_.model.history
y_pred = grid.predict(x_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
metrics = history.history
metrics.keys()
plt.figure(figsize=(14,7))

plt.plot(metrics['acc'])
plt.plot(metrics['loss'])
plt.title('Curvas de Treinamento')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Precisão', 'Função de Custo'], loc='upper left')
plt.show();
```

```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import torch
import torch.utils.data

def load_data(train_path, test_path):
    train_dataset = h5py.File(train_path,'r')
    train_set_X = np.array(train_dataset['train_set_x'][:])
    train_set_Y = np.array(train_dataset['train_set_y'][:])
    
    test_dataset = h5py.File(test_path,'r')
    test_set_X = np.array(test_dataset['test_set_x'][:])
    test_set_Y = np.array(test_dataset['test_set_y'][:])
    
    classes = np.array(test_dataset['list_classes'][:])    
    return train_set_X, train_set_Y, test_set_X, test_set_Y, classes

def reshape_data(x_dataset, y_dataset):
    x_dataset_reshape = x_dataset.reshape((x_dataset.shape[0], x_dataset.shape[1] * x_dataset.shape[2] * x_dataset.shape[3]))
    y_dataset_reshape = y_dataset.reshape((y_dataset.shape[0],-1))
    return x_dataset_reshape, y_dataset_reshape

def normalize(imgs):
    return imgs / 255.

def convert_to_tensor(numpy_array):
    data = normalize(numpy_array)
    return torch.from_numpy(data)
```

```python
class ConvNet(nn.Module):
  
  def __init__(self):
    super(ConvNet, self).__init__()
    self.layer1 = nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=16, kernel_size=5, stride=2, padding=1),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=5, stride=2))

    self.layer2 = nn.Sequential(
        nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=2, padding=1),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=3, stride=2))

    self.fc1 = nn.Linear(12288,4)
    self.fc2 = nn.Linear(4,3)
    self.fc3 = nn.Linear(3,1)

  def forward(self, x):
      x = self.layer1(x)
      x = self.layer2(x)

      x = x.view(x.shape[0], -1)
#       print(x.shape)
      x = self.fc1(x)
      x = self.fc2(x)
      x = self.fc3(x)
    
      return F.log_softmax(x, dim=1)
        
```
